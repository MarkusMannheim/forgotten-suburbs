```python
import pandas as pd

demographics = pd.read_csv("demographics.csv", index_col="suburb", thousands=",")
demographics.income = demographics.income.apply(lambda x: int(x.replace("$", "").replace(",", "")))
sales = pd.read_csv("sale_data.csv")

sales = sales[sales.type.apply(lambda x: True if ("Single" in str(x)) | (str(x) == "Residence") else False)]
sales = sales[["suburb", "year", "price"]]
sales.price = sales.price.apply(lambda x: x.replace(",", "").replace("$", ""))
sales = sales[sales.price.apply(lambda x: (x.isnumeric()) & (x != "0"))]
sales.price = sales.price.astype("int")

for suburb in demographics.index:
    data_2007 = sales[(sales.suburb == suburb) & (sales.year == 2007)]
    data_2020 = sales[(sales.suburb == suburb) & (sales.year == 2020)]
    demographics.loc[suburb, "house_price_2007"] = data_2007.price.median()
    demographics.loc[suburb, "house_price_2020"] = data_2020.price.median()

demographics.dropna(inplace=True)
demographics["pop_growth"] = demographics["pop_now"] / demographics["pop_before"] - 1
demographics["price_growth"] = demographics["house_price_2020"] / demographics["house_price_2007"] - 1
demographics.columns = ["state", "household_income", "population_2006", "population_2016", "house_price_2007", "house_price_2020", "population_growth", "house_price_growth"]
demographics.to_csv("forgotten_suburbs.csv")
```
